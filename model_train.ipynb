```python
import numpy as np
#extracting data from csv files into numpy arrays
from numpy import genfromtxt
x_train = genfromtxt('train_data.csv', delimiter=',')
y_train = genfromtxt('train_labels.csv', delimiter=',')
x_test = genfromtxt('test_data.csv', delimiter=',')
y_test = genfromtxt('test_labels.csv', delimiter=',')
```

```python
import random
tmp = [[x,y] for x, y in zip(x_train, y_train)]
tmp2 = [[x,y] for x, y in zip(x_test, y_test)]
random.shuffle(tmp)
random.shuffle(tmp2)
x_train = [n[0] for n in tmp]
y_train = [n[1] for n in tmp]
x_test = [n[0] for n in tmp2]
y_test = [n[1] for n in tmp2]
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
x_train.shape,y_train.shape
```

```python
#converting to one hot
from keras.utils.np_utils import to_categorical
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
y_train.shape,y_test.shape
```

```python
#reshaping to 2D 
x_train=np.reshape(x_train,(x_train.shape[0], 40,1))
x_test=np.reshape(x_test,(x_test.shape[0], 40,1))
x_train.shape,x_test.shape
```

```python
#reshaping to shape required by CNN
x_train=np.reshape(x_train,(x_train.shape[0], 40,1,1))
x_test=np.reshape(x_test,(x_test.shape[0], 40,1,1))
```

```python
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout
```

```python
#forming model
model=Sequential()
```

```python
model.add(Conv2D(32,(3*3),padding="Same",activation="relu",input_shape=(40,1,1)))
model.add(MaxPooling2D(padding="same"))

model.add(Conv2D(64,(3*3),padding="same",activation="relu"))
model.add(MaxPooling2D(padding="same"))
model.add(Dropout(0.25))

model.add(Conv2D(128,(3*3),padding="same",activation="relu"))
model.add(MaxPooling2D(padding="same"))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(3,activation="softmax"))
```

```python
#compiling
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
#training the model
model.fit(x_train,y_train,batch_size=5,epochs=20,validation_data=(x_test,y_test))
```

```python
model.save('outside_model3.h5')
```

```python
#train and test loss and scores respectively
train_loss_score=model.evaluate(x_train,y_train)
test_loss_score=model.evaluate(x_test,y_test)
print(train_loss_score)
print(test_loss_score)
```
